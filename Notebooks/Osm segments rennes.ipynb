```python
import folium
from folium.plugins import FastMarkerCluster
from folium.plugins import MarkerCluster
from pymongo import MongoClient
import pandas as pd
import numpy as np
```

```python
client = MongoClient()
osm = client["osm-rennes"]
osmWays= osm.roads

cur = osmWays.find(   
    {
        "loc": {
            "$geoIntersects": {
                "$geometry": {
                    "type": "Polygon",
                    "coordinates": [
                        [[-1.5460, 48.1656],
                         [-1.5460, 48.0632],
                         [-1.7626, 48.0632],
                         [-1.7626,48.1656],
                         [-1.5460, 48.1656]]
                    ]
                }
            }
        },    
        "tag.k": "highway",    
        "tag.v": {
          "$in": [
            "motorway", 
            "trunk", 
            "primary",
            "secondary", 
            "tertiary", 
            "motorway_link", 
            "trunk_link", 
            "primary_link",
            "secondary_link", 
            "tertiary_link"
          ]
        }
    },
    {'_id': 1}
)

main_roads = [r['_id'] for r in cur]
main_roads
```
